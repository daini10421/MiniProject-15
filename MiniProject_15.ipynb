# **"Convert 100 images to Grey Scale (b/w)" using OpenCV**

```python
import cv2
import os
import requests
from urllib.parse import urlparse

# GitHub repository URL
repo_url = "https://github.com/daini10421/100images"

# Iterate over 100 images
for i in range(1, 101):
    # Create the URL to the raw image file
    image_url = f"{repo_url}/raw/main/image({i}).png"

    # Parse the URL
    parsed_url = urlparse(image_url)

    # Fetch the image filename from the URL
    image_filename = os.path.basename(parsed_url.path)

    try:
        # Download the image using requests
        response = requests.get(image_url)
        with open(image_filename, 'wb') as f:
            f.write(response.content)

        # Load the input image
        img = cv2.imread(image_filename)

        if img is not None:
            # Obtain the dimensions of the image array
            (row, col) = img.shape[0:2]

            # Take the average of pixel values of the BGR Channels
            # to convert the colored image to grayscale image
            for r in range(row):
                for c in range(col):
                    # Find the average of the BGR pixel values
                    img[r, c] = sum(img[r, c]) * 0.33

            # Show the grayscale image
            cv2.imshow('Grayscale Image', img)
            cv2.waitKey(0)

            # Remove the downloaded image
            os.remove(image_filename)
        else:
            print(f" Converted  {image_filename}  to greyscale")

    except Exception as e:
        print(f" downloading image: {image_url}")
        print(f"Error message: {str(e)}")

# Close all windows
cv2.destroyAllWindows()
```
